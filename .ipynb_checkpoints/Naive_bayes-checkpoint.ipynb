```python
import numpy as np
import xml.etree.ElementTree as ET
from IPython.display import display
import matplotlib.pyplot as plt
import sklearn
import os

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
```

```python
import nltk
nltk.download('punkt')
```

```python
stopword = set(stopwords.words('english'))
porter = PorterStemmer()
```

```python
token_dict = {}
for file in os.listdir('./src/train/find_pmid_with_abstract'):
    et = ET.parse('./src/train/find_pmid_with_abstract/'+file)
    root = et.getroot()
    article = root.find('MedlineCitation').find('Article').findall('Abstract')
    for a in article:
        atext = a.find('AbstractText').text
        if type(atext) == str:
            tokens = word_tokenize(atext)
            words = [word.lower() for word in tokens if word.isalpha()]
            clean_word = [word for word in words if not word in stopword]
            for w in clean_word:
                if not w in token_dict:
                    token_dict[w] = 1
                else:
                    token_dict[w] += 1
        else:
            print(atext)
print(token_dict)
```

```python
token_sorted = sorted(token_dict.items(), key=lambda kv: kv[1],reverse=True)
```

```python
token_word = {}
for token in token_sorted:
    token_word[token[0]] = token[1]
```

```python
import string

letters = list(string.ascii_lowercase)
char_map = dict(zip(letters, range(len(letters))))
```

```python
def  count_strcount_st (string):
    global letters, char_map
    result = np.zeros(len(letters))
    np_str = np.array(list(string))
    str_char, str_char_count = np.unique(np_str, return_counts=True)
    for char, count in zip(str_char, str_char_count):
        result[char_map[char]] = count
    return result
```
