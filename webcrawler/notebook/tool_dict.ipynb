```python
from urllib.request import urlopen,Request 
from bs4 import BeautifulSoup 
import requests
import time
import os
from tqdm import tqdm
import pickle
import pandas as pd
```

```python
def generate_category_dict():
        state = ""
        category_dict = {}
        tool_file = open("../omictool_toollist.txt","r")
        for t in tool_file:
                t_list = t.strip().split()
                for i in range(1,len(t_list)-1,2):
                    category_dict[t_list[i]] = t_list[i+1]
        return category_dict
```

```python
def generate_category_lookup():
        state = ""
        category_tree = {}
        tool_file = open("../omictool_toollist.txt","r")
        for t in tool_file:
                t_list = t.strip().split()
                if t_list[-3] not in category_tree:
                    if len(t_list) > 5:
                        category_tree[t_list[-3]] = t_list[-5] 
                    else:
                        category_tree[t_list[-3]] = None
                        
                    
        return category_tree
```

```python
def generate_tool_dict(category_lookup,interested_category):
        state = ""
        tool_dict = {}
        try:
                tool_file = open("../omictool_toollist.txt","r")
                for t in tool_file:
                        tool = t.strip().split()[-1]
                        state = tool
                        print("\r Send Request",end="")
                        omictool = "https://omictools.com"+tool
                        response = requests.get(omictool)
                        print("\r Scraping "+omictool,end=" ")
                        html = BeautifulSoup(response.text,"html5lib")
                        a_list = html.find_all('a')
                        if tool not in tool_dict:
                            tool_dict[tool] = set()
#                         print(a_list)
                        for a in a_list:
#                                 print(a.get('href'))
                                if a.get('href') != None:
                                    if a.get('href')[-8:] == "category":
                                        if a.get('href') in interested_category:
                                            tool_dict[tool].add(a.get('href'))
                                        else:
                                            if a.get('href') in category_lookup:
                                                tool_dict[tool].add(category_lookup[a.get('href')])
#                         print(tool_dict)
                        time.sleep(0.1)
                return tool_dict
        except:
            raise
```

```python
def generate_pmid_dict():
    state = ""
    pmid_dict = {}
    try:
            tool_file = open("../omictool_toollist_pmid.txt","r")
            for t in tool_file:
                    pmid = t.strip().split()[-1]
                    tool = t.strip().split()[-2]
                    if pmid not in pmid_dict:
                        pmid_dict[pmid] = set()
                    pmid_dict[pmid].add(tool)
            return pmid_dict
    except:
        raise
```

```python
def pmid_mapping(tool_dict,pmid_dict):
    mapping = {}
    for p in pmid_dict:
        if p not in mapping:
            mapping[p] = set()
        for tool in pmid_dict[p]:
            for t in tool_dict[tool]:
                mapping[p].add(t)
    return mapping
```

```python
category_dict = generate_category_dict()
category_lookup = generate_category_lookup()
```

```python
category_tree = {}
for i in category_lookup:
    if category_lookup[i] != None:
        if category_lookup[i] not in category_tree:
            category_tree[category_lookup[i]] = set()
        category_tree[category_lookup[i]].add(i)
    else:
        category_tree[i] = None
```

```python
interested_category = set()
for i in category_tree:
    if category_tree[i] != None:
        number_list = [int(category_dict[j]) for j in category_tree[i]]
        checking = all(j > 100 for j in number_list)
        if checking:
            for child in category_tree[i]:
                interested_category.add(child)
        else:
            interested_category.add(i)
```

```python
tool_dict = generate_tool_dict(category_lookup,interested_category)
```
