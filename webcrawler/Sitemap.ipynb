```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from graphviz import Digraph
```

```python
maintree = {}
subtree = {}
```

```python
main_cate = open("omictool_maincategory.txt","r")
sub_cate = open("omictool_subcategory.txt","r")
```

```python
# node = set()
for line in main_cate:
    link = line.split()
    if link[0] not in maintree:
        maintree[link[0]] = set()
    maintree[link[0]].add((link[1],link[2]))
```

```python
# node = set()
for line in sub_cate:
    link = line.split()
    print(link)
    if (link[0],link[1],link[2]) not in subtree:
#         print()
        subtree[(link[0],link[1],link[2])] = set()
    if len(link) > 4:
        subtree[(link[0],link[1],link[2])].add((link[3],link[4]))
    else:
        subtree[(link[0],link[1],link[2])].add(' ')
        
```

```python
subtree
```

```python
tree = {}
for node in maintree:
    print(node)
    if node not in tree:
        tree[node] = {}
    for e in maintree[node]:
        if e not in tree[node]:
            tree[node][e] = []
```

```python
for node in subtree:
    for e in subtree[node]:
        print(node,e)
        tree[node[0]][(node[1],node[2])].append(e)
```

```python
data_list = []
for p in tree:
    for c in tree[p]:
        print('c',c)
        if len(tree[p][c]) == 0:
            li = [p,c[0],c[1],' ',' ']
            data_list.append(li)
        for i in tree[p][c]:
            print('i',i)
            if i != ' ':
                li = [p,c[0],c[1],i[0],i[1]]
                data_list.append(li)
            else:
                li = [p,c[0],c[1],' ',' ']
                data_list.append(li)
                
```

```python
d = pd.DataFrame(data_list,columns=['Main','Sub1','number of tools 1','Sub2','number of tools 2'])
```

```python
d = d.set_index(['Main','Sub1','number of tools 1','Sub2'])
```
