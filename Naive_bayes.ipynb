```python
import numpy as np
import xml.etree.ElementTree as ET
from IPython.display import display
import matplotlib.pyplot as plt
import sklearn
import os

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
```

```python
stopword = set(stopwords.words('english'))
porter = PorterStemmer()
```

```python
pos_abstract = []
i = 0
dir_leng = len(os.listdir('./src/train/find_pmid_with_abstract'))
print(dir_leng)
for file in os.listdir('./src/train/find_pmid_with_abstract'):
    i += 1
    print(str((i/dir_leng)*100),end="\r")
    et = ET.parse('./src/train/find_pmid_with_abstract/'+file)
    root = et.getroot()
    article = root.find('MedlineCitation').find('Article').findall('Abstract')
    word = {}
    for a in article:
        atext = a.find('AbstractText').text
        if type(atext) == str:
            pos_abstract.append(atext)
```

```python
token_dict = {}
pos_words_list = []
pos_abs_leng = len(pos_abstract)
for abst in pos_abstract:
    pos_words = {}
    tokens = word_tokenize(abst)
    words = [word.lower() for word in tokens if word.isalpha()]
    clean_word = [word for word in words if not word in stopword]
#     print(clean_word)
    for w in clean_word:
        if not w in token_dict:
            token_dict[w] = 1
        else:
            token_dict[w] += 1
        pos_words[w] = True
#     print(pos_words)
    pos_words_list.append((pos_words,'pos'))
print(token_dict)
```

```python
token_sorted = sorted(token_dict.items(), key=lambda kv: kv[1],reverse=True)
```

```python
token_word = {}
for token in token_sorted:
    token_word[token[0]] = token[1]
```

```python
neg_abstract = []
i = 0
negdir_leng = len(os.listdir('./src/train/nontool_with_abstract'))
print(negdir_leng)
for file in os.listdir('./src/train/nontool_with_abstract'):
    i += 1
    print(str((i/negdir_leng)*100),end="\r")
    et = ET.parse('./src/train/nontool_with_abstract/'+file)
    root = et.getroot()
    article = root.find('MedlineCitation').find('Article').findall('Abstract')
    word = {}
    for a in article:
        atext = a.find('AbstractText').text
        if type(atext) == str:
            neg_abstract.append(atext)
```

```python
neg_token_dict = {}
neg_words_list = []
neg_abs_leng = len(pos_abstract)
for neg in neg_abstract:
    neg_words = {}
    tokens = word_tokenize(neg)
    words = [word.lower() for word in tokens if word.isalpha()]
    clean_word = [word for word in words if not word in stopword]
    for w in clean_word:
        if not w in neg_token_dict:
            neg_token_dict[w] = 1
        else:
            neg_token_dict[w] += 1
        neg_words[w] = True 
    neg_words_list.append((neg_words,'neg'))
```

```python
import random
negcutoff = int(len(neg_words_list)*(4/5))
poscutoff = int(len(pos_words_list)*(4/5))

print(poscutoff,negcutoff,len(pos_words_list),len(neg_words_list))

random.shuffle(neg_words_list)
random.shuffle(pos_words_list)

train_fic = neg_words_list[:negcutoff] + pos_words_list[:poscutoff]
test_fic = neg_words_list[negcutoff:] + pos_words_list[poscutoff:]

print('train on '+str(len(train_fic))+' docs, test on '+str(len(test_fic))+ ' docs')
```
